```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import skew
from scipy.stats.stats import pearsonr
import matplotlib
```

```python
plt.rcParams['figure.figsize'] = (12,8)
```

```python
raw_file_train=pd.read_csv('train.csv')
```

```python
raw_file_test=pd.read_csv('test.csv')
```

```python
raw_file_train
```

```python
all_data=pd.concat((raw_file_train.loc[:,'MSSubClass':'SaleCondition'],raw_file_test.loc[:,'MSSubClass':'SaleCondition']))
```

```python
all_data
```

```python
prices = pd.DataFrame({"price":raw_file_train["SalePrice"], "log(price+1)":np.log1p(raw_file_train["SalePrice"])})
```

```python
prices.hist()
```

```python
raw_file_train['SalePrice']=np.log1p(raw_file_train['SalePrice'])
```

```python
numeric_feats=all_data.dtypes[all_data.dtypes !='object'].index
```

```python
numeric_feats
```

```python
skewed_feats=raw_file_train[numeric_feats].apply(lambda x: skew(x.dropna()))
```

```python
skewed_feats=skewed_feats[skewed_feats>0.75]
skewed_feats=skewed_feats.index
```

```python
all_data[skewed_feats]=np.log1p(all_data[skewed_feats])
```

```python
all_data=pd.get_dummies(all_data)
```

```python
all_data
```

```python
all_data=all_data.fillna(all_data.mean())
```

```python
X_train=all_data[:raw_file_train.shape[0]]
X_test=all_data[raw_file_train.shape[0]:]
y=raw_file_train['SalePrice']
```

Model Implementation:

```python
from sklearn.linear_model import Ridge, RidgeCV, ElasticNet, LassoCV, LassoLarsCV 
#importing necessary libraries for implementing LASSO(l1) and Ridge(l2) Regression
from sklearn.model_selection import cross_val_score
```

```python
def rmse_cv(model):
    rmse=np.sqrt(-cross_val_score(model,X_train,y,scoring="neg_mean_squared_error",cv=5))
    return(rmse)
```

```python
model_ridge=Ridge()
```

```python
alphas=[0.05,0.1,0.3,1,3,5,10,15,30,50,75]
cv_ridge=[rmse_cv(Ridge(alpha=alpha)).mean()
             for alpha in alphas]
```

```python
cv_ridge=pd.Series(cv_ridge, index=alphas)
cv_ridge.plot(title='Validation - Parameter picking')
plt.xlabel('alpha')
plt.ylabel('rmse')
```

```python
cv_ridge.min()
```

```python
model_lasso=LassoCV(alphas=[1,0.1,0.001,0.0005]).fit(X_train,y)
```

```python
rmse_cv(model_lasso).mean()
```

```python
# Lasso regression is showing better results when compared to Ridge Regression. 
```

```python
coef=pd.Series(model_lasso.coef_, index=X_train.columns)
```

```python
print(str(sum(coef !=0))+" Variables picked from Lasso Regression." + str(sum(coef ==0))+" Variables left out")
```

```python
imp_coef=pd.concat([coef.sort_values().head(10),coef.sort_values().tail(10)])
```

```python
imp_coef.plot(kind="bar", title="Selected Coefficients from Lasso")
```

```python
matplotlib.rcParams['figure.figsize']=(10.0,10.0)
preds=pd.DataFrame({"preds":model_lasso.predict(X_train),"true":y}) #Looking at the predictions!!
```

```python
preds['residuals']=preds['true']-preds['preds']
```

```python
preds.plot(x="preds",y="residuals",kind="scatter")
```

```python
model_lasso.predict(X_test)
```

```python
test_set_predictions=model_lasso.predict(X_test)
```

```python
from keras.layers import Dense
from keras.models import Sequential
from keras.regularizers import l1
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train = StandardScaler().fit_transform(X_train)
```

```python
X_tr,X_val,y_tr,y_val=train_test_split(X_train,y,random_state=5)
```

```python
model=Sequential()
model.add(Dense(1,input_dim=X_train.shape[1],W_regularizer=l1(0.001)))
model.compile(loss="mse",optimizer="adam")
```

```python
model.summary()
```

```python
hist=model.fit(X_tr, y_tr, validation_data=(X_val,y_val))
```

```python
pd.Series(model.predict(X_val)[:,0]).hist()
```
